```python
def RMSE(a, b):
    mse = ((a - b) ** 2).mean()  # 두 값의 차이의 제곱의 평균
    rmse = mse ** 0.5        # MSE의 제곱근
    return rmse
print("맥북!")
```

```python
x = macbook["used_years"].values
y = macbook["price"].values

predictions = model(x, w, b)
print(predictions)
```

```python
rmse = RMSE(predictions, y)
rmse
```

```python
def loss(x, w, b, y):
    predictions = model(x, w, b)
    L = RMSE(predictions, y)
    return L
print("맥북!")
```

```python
LEARNING_RATE = 1
print("맥북!")
```

```python
x = macbook["used_years"].values
y = macbook["price"].values
print("맥북!")
```

```python
w = 3.1
b = 2.3
w, b
```

```python
losses = []
print("맥북1")
```

```python
%%time
def gradient(x, w, b, y):
    N = len(x)
    predictions = model(x, w, b)
    error = predictions - y

    # RMSE가 0이 되는 경우를 방지하기 위해 작은 값을 더해줍니다.
    current_rmse = RMSE(predictions, y) + 1e-8

    dw = (1 / (N * current_rmse)) * np.sum(error * x)
    db = (1 / (N * current_rmse)) * np.sum(error)

    return dw, db

for i in range(1, 2001):
    dw, db = gradient(x, w, b, y)   # 3, 4번: 모델이 prediction을 예측하고, 손실함수값을 계산함과 동시에 기울기 계산
    w -= LEARNING_RATE * dw         # 5번: w = w - η * dw 로 업데이트
    b -= LEARNING_RATE * db         # 5번: b = b - η * db 로 업데이트
    L = loss(x, w, b, y)            # 현재의 loss 값 계산
    losses.append(L)                # loss 값 기록
    if i % 100 == 0:
        print('Iteration %d : Loss %0.4f' % (i, L))
```

```python
plt.plot(losses)
plt.show()
```
