```python
# 사각형 넓이를 구하는 클래스
class Square:
    def __init__(self):
        self.square = int(input(
            '넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n'
            '1.직사각형 2.평행사변형 3.사다리꼴\n >>> '
        ))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')
        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    # 직사각형 넓이
    def rect(self):
        width, vertical = map(
            int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>> ').split(',')
        )
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    # 평행사변형 넓이
    def par(self):
        base, height = map(
            int, input('밑변, 높이를 입력하세요. 예시 : 밑변,높이\n >>> ').split(',')
        )
        area = base * height
        result = '평행사변형의 넓이는 : ' + str(area)
        return result

    # 사다리꼴 넓이
    def trape(self):
        base, top, height = map(
            int, input('밑변, 윗변, 높이를 입력하세요. 예시 : 밑변,윗변,높이\n >>> ').split(',')
        )
        area = (base + top) * height // 2
        result = '사다리꼴의 넓이는 : ' + str(area)
        return result


# 실행 예시
a = Square()
```

```python
# 키오스크 클래스
class Kiosk:
    def __init__(self):
        # 메뉴와 가격
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]

        # 주문 내역
        self.order_menu = []
        self.order_price = []

    # 메뉴 출력
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 및 추가 주문
    def menu_select(self):
        n = 1
        while n != 0:
            n = int(input("음료 번호를 입력하세요 (주문 종료는 0) : "))

            if n == 0:
                break

            if 1 <= n <= len(self.menu):
                # 가격
                self.price_sum = self.price[n - 1]

                # 온도 선택
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = 'HOT'
                    elif t == 2:
                        self.temp = 'ICE'
                    else:
                        print("1과 2 중 하나를 입력하세요.")

                # 주문 리스트 저장
                self.order_menu.append(self.temp + ' ' + self.menu[n - 1])
                self.order_price.append(self.price_sum)

                print('추가 주문을 원하시면 계속 입력해주세요.')
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

    # 지불
    def pay(self):
        print('\n총 결제 금액은', sum(self.order_price), '원입니다.')

        p = 0
        while p != 1 and p != 2:
            p = int(input("결제 방법을 선택하세요. 현금은 1번, 카드는 2번입니다 : "))
            if p == 1:
                print('직원을 호출하겠습니다.')
            elif p == 2:
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
            else:
                print('다시 결제를 시도해 주세요.')

    # 주문표 출력
    def table(self):
        print('⟝' + '-' * 30 + '⟞')
        for _ in range(5):
            print('|' + ' ' * 31 + '|')

        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i], '원')

        print('합계 금액 :', sum(self.order_price), '원')

        for _ in range(5):
            print('|' + ' ' * 31 + '|')
        print('⟝' + '-' * 30 + '⟞')


# 실행 예시
a = Kiosk()
a.menu_print()
a.menu_select()
a.pay()
a.table()
```
