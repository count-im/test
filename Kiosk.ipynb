```python
# ì‚¬ê°í˜• ë„“ì´ë¥¼ êµ¬í•˜ëŠ” í´ë˜ìŠ¤
class Square:
    def __init__(self):
        self.square = int(input(
            'ë„“ì´ë¥¼ êµ¬í•˜ê³  ì‹¶ì€ ì‚¬ê°í˜•ì˜ ìˆ«ìë¥¼ ì¨ì£¼ì„¸ìš”.\n'
            '1.ì§ì‚¬ê°í˜• 2.í‰í–‰ì‚¬ë³€í˜• 3.ì‚¬ë‹¤ë¦¬ê¼´\n >>> '
        ))

        if self.square == 1:
            print('ì§ì‚¬ê°í˜• í•¨ìˆ˜ëŠ” rect()ì…ë‹ˆë‹¤.')
        elif self.square == 2:
            print('í‰í–‰ì‚¬ë³€í˜• í•¨ìˆ˜ëŠ” par()ì…ë‹ˆë‹¤.')
        elif self.square == 3:
            print('ì‚¬ë‹¤ë¦¬ê¼´ í•¨ìˆ˜ëŠ” trape()ì…ë‹ˆë‹¤.')
        else:
            print('1, 2, 3 ì¤‘ì—ì„œ ë‹¤ì‹œ ì…ë ¥í•´ì£¼ì„¸ìš”')

    # ì§ì‚¬ê°í˜• ë„“ì´
    def rect(self):
        width, vertical = map(
            int, input('ê°€ë¡œ, ì„¸ë¡œë¥¼ ì…ë ¥í•˜ì„¸ìš”. ì˜ˆì‹œ : ê°€ë¡œ,ì„¸ë¡œ\n >>> ').split(',')
        )
        area = width * vertical
        result = 'ì§ì‚¬ê°í˜•ì˜ ë„“ì´ëŠ” : ' + str(area)
        return result

    # í‰í–‰ì‚¬ë³€í˜• ë„“ì´
    def par(self):
        base, height = map(
            int, input('ë°‘ë³€, ë†’ì´ë¥¼ ì…ë ¥í•˜ì„¸ìš”. ì˜ˆì‹œ : ë°‘ë³€,ë†’ì´\n >>> ').split(',')
        )
        area = base * height
        result = 'í‰í–‰ì‚¬ë³€í˜•ì˜ ë„“ì´ëŠ” : ' + str(area)
        return result

    # ì‚¬ë‹¤ë¦¬ê¼´ ë„“ì´
    def trape(self):
        base, top, height = map(
            int, input('ë°‘ë³€, ìœ—ë³€, ë†’ì´ë¥¼ ì…ë ¥í•˜ì„¸ìš”. ì˜ˆì‹œ : ë°‘ë³€,ìœ—ë³€,ë†’ì´\n >>> ').split(',')
        )
        area = (base + top) * height // 2
        result = 'ì‚¬ë‹¤ë¦¬ê¼´ì˜ ë„“ì´ëŠ” : ' + str(area)
        return result


# ì‹¤í–‰ ì˜ˆì‹œ
a = Square()
```

```python
# í‚¤ì˜¤ìŠ¤í¬ í´ë˜ìŠ¤
class Kiosk:
    def __init__(self):
        # ë©”ë‰´ì™€ ê°€ê²©
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]

        # ì£¼ë¬¸ ë‚´ì—­
        self.order_menu = []
        self.order_price = []

    # ë©”ë‰´ ì¶œë ¥
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], 'ì›')

    # ì£¼ë¬¸ ë° ì¶”ê°€ ì£¼ë¬¸
    def menu_select(self):
        n = 1
        while n != 0:
            n = int(input("ìŒë£Œ ë²ˆí˜¸ë¥¼ ì…ë ¥í•˜ì„¸ìš” (ì£¼ë¬¸ ì¢…ë£ŒëŠ” 0) : "))

            if n == 0:
                break

            if 1 <= n <= len(self.menu):
                # ê°€ê²©
                self.price_sum = self.price[n - 1]

                # ì˜¨ë„ ì„ íƒ
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT ìŒë£ŒëŠ” 1ì„, ICE ìŒë£ŒëŠ” 2ë¥¼ ì…ë ¥í•˜ì„¸ìš” : "))
                    if t == 1:
                        self.temp = 'HOT'
                    elif t == 2:
                        self.temp = 'ICE'
                    else:
                        print("1ê³¼ 2 ì¤‘ í•˜ë‚˜ë¥¼ ì…ë ¥í•˜ì„¸ìš”.")

                # ì£¼ë¬¸ ë¦¬ìŠ¤íŠ¸ ì €ì¥
                self.order_menu.append(self.temp + ' ' + self.menu[n - 1])
                self.order_price.append(self.price_sum)

                print('ì¶”ê°€ ì£¼ë¬¸ì„ ì›í•˜ì‹œë©´ ê³„ì† ì…ë ¥í•´ì£¼ì„¸ìš”.')
            else:
                print("ì—†ëŠ” ë©”ë‰´ì…ë‹ˆë‹¤. ë‹¤ì‹œ ì£¼ë¬¸í•´ ì£¼ì„¸ìš”.")

    # ì§€ë¶ˆ
    def pay(self):
        print('\nì´ ê²°ì œ ê¸ˆì•¡ì€', sum(self.order_price), 'ì›ì…ë‹ˆë‹¤.')

        p = 0
        while p != 1 and p != 2:
            p = int(input("ê²°ì œ ë°©ë²•ì„ ì„ íƒí•˜ì„¸ìš”. í˜„ê¸ˆì€ 1ë²ˆ, ì¹´ë“œëŠ” 2ë²ˆì…ë‹ˆë‹¤ : "))
            if p == 1:
                print('ì§ì›ì„ í˜¸ì¶œí•˜ê² ìŠµë‹ˆë‹¤.')
            elif p == 2:
                print('ICì¹© ë°©í–¥ì— ë§ê²Œ ì¹´ë“œë¥¼ ê½‚ì•„ì£¼ì„¸ìš”.')
            else:
                print('ë‹¤ì‹œ ê²°ì œë¥¼ ì‹œë„í•´ ì£¼ì„¸ìš”.')

    # ì£¼ë¬¸í‘œ ì¶œë ¥
    def table(self):
        print('âŸ' + '-' * 30 + 'âŸ')
        for _ in range(5):
            print('|' + ' ' * 31 + '|')

        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i], 'ì›')

        print('í•©ê³„ ê¸ˆì•¡ :', sum(self.order_price), 'ì›')

        for _ in range(5):
            print('|' + ' ' * 31 + '|')
        print('âŸ' + '-' * 30 + 'âŸ')


# ì‹¤í–‰ ì˜ˆì‹œ
a = Kiosk()
a.menu_print()
a.menu_select()
a.pay()
a.table()
```
